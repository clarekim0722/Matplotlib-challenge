```python
# 0. Three Observations 

# 1) There is a strong positive correlation between the mouse weight and average tumor volume. 
# 2) When ingested with drug 'Capomulin', the average tumor volume decreased over time, which indicates the efficiency of the drug. 
# 3) Treatments like 'Capomulin' and 'Ramicane' seem to be more effective than treatments like 'Infubinol' and 'Ceftamin' since the former two resulted in higher tumor volume decrease. 
```

```python
# 1. Prepare the Data 

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
study_results = pd.read_csv('Study_results.csv')
mouse_metadata = pd.read_csv('Mouse_metadata.csv')

# Merge mouse_metadata and study_results into a single DataFrame
merged_data = pd.merge(mouse_metadata, study_results, on='Mouse ID')

# Display the number of unique mice IDs
unique_mice_ids_before = merged_data['Mouse ID'].nunique()
print("Number of unique mice IDs before cleaning:", unique_mice_ids_before)
```

```python
# Check for any mouse ID with duplicate time points
duplicate_mouse = merged_data[merged_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]

# Display the data associated with that mouse ID
if not duplicate_mouse.empty:
    print("Data associated with the duplicate mouse ID:")
    print(duplicate_mouse)
else:
    print("No duplicate mouse IDs found.")

# Create a new DataFrame where this data is removed
cleaned_data = merged_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep=False)

# Display the updated number of unique mice IDs
unique_mice_ids_after = cleaned_data['Mouse ID'].nunique()
print("Updated number of unique mice IDs after cleaning:", unique_mice_ids_after)
```

```python
#2. Generate Summary Statistics 
# Group the data by Drug and calculate summary statistics
summary_statistics = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Display the summary statistics DataFrame
print("Summary Statistics for Tumor Volume by Drug Regimen:")
print(summary_statistics)
```

```python
#3. Create Bar Charts and Pie Charts

# Calculate the total number of rows (Mouse ID/Timepoints) for each drug regimen
total_rows_per_drug = cleaned_data['Drug Regimen'].value_counts()

# First bar chart using Pandas DataFrame.plot() method
total_rows_per_drug.plot(kind='bar', color='skyblue', alpha=0.7, edgecolor='black', figsize=(10, 6))
plt.title('Total Rows per Drug Regimen (Pandas Plot)')
plt.xlabel('Drug Regimen')
plt.ylabel('Total Rows (Mouse ID/Timepoints)')
plt.xticks(rotation=45)
plt.show()

# Second bar chart using Matplotlib's pyplot methods
plt.figure(figsize=(10, 6))
plt.bar(total_rows_per_drug.index, total_rows_per_drug, color='skyblue', alpha=0.7, edgecolor='black')
plt.title('Total Rows per Drug Regimen (Matplotlib)')
plt.xlabel('Drug Regimen')
plt.ylabel('Total Rows (Mouse ID/Timepoints)')
plt.xticks(rotation=45)
plt.show()

# Calculate the distribution of female versus male mice
gender_distribution = cleaned_data['Sex'].value_counts()

# First pie chart using Pandas DataFrame.plot() method
gender_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'],
                          figsize=(8, 8), explode=(0.1, 0), shadow=True, startangle=140)
plt.title('Distribution of Female vs Male Mice (Pandas Plot)')
plt.ylabel('')  # Remove the 'Sex' label
plt.show()

# Second pie chart using Matplotlib's pyplot methods
plt.figure(figsize=(8, 8))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%',
        colors=['lightcoral', 'lightskyblue'], explode=(0.1, 0), shadow=True, startangle=140)
plt.title('Distribution of Female vs Male Mice (Matplotlib)')
plt.show()
```

```python
#4. Calculate Quartiles, Find Outliers, and Create a Box Plot
# Step 1: Create a grouped DataFrame for the last time point of each mouse
last_timepoint_per_mouse = cleaned_data.groupby('Mouse ID')['Timepoint'].max()
last_timepoint_data = pd.DataFrame(last_timepoint_per_mouse).reset_index()

# Merge the grouped DataFrame with the original cleaned DataFrame to get the final tumor volume data
final_tumor_volume_data = pd.merge(last_timepoint_data, cleaned_data, on=['Mouse ID', 'Timepoint'])

# Step 2: Create a list to hold treatment names and an empty list for tumor volume data
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_volume_data = []

# Step 3: Loop through each drug, locate the rows, and append the final tumor volumes
for drug in treatments:
    drug_data = final_tumor_volume_data[final_tumor_volume_data['Drug Regimen'] == drug]['Tumor Volume (mm3)']
    tumor_volume_data.append(drug_data)

# Step 4: Calculate quartiles and IQR, determine outliers, and print the results
for i in range(len(treatments)):
    quartiles = tumor_volume_data[i].quantile([0.25, 0.5, 0.75])
    lower_quartile = quartiles[0.25]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    lower_bound = lower_quartile - (1.5 * iqr)
    upper_bound = upper_quartile + (1.5 * iqr)

    outliers = tumor_volume_data[i][(tumor_volume_data[i] < lower_bound) | (tumor_volume_data[i] > upper_bound)]

    print(f"Results for {treatments[i]}:")
    print(f"  Lower Quartile: {lower_quartile}")
    print(f"  Upper Quartile: {upper_quartile}")
    print(f"  Interquartile Range (IQR): {iqr}")
    print(f"  Lower Bound: {lower_bound}")
    print(f"  Upper Bound: {upper_bound}")
    print(f"  Potential Outliers: {outliers.tolist()}")
    print("\n")
    
# Step 5: Generate a box plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot(tumor_volume_data, labels=treatments, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 8})
plt.title('Distribution of Final Tumor Volume for Promising Treatment Regimens')
plt.xlabel('Treatment Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
#5. Create a Line Plot and a Scatter Plot

# Step 1: Select a single mouse treated with Capomulin
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']
selected_mouse = capomulin_data['Mouse ID'].sample(n=1).iloc[0]

# Filter data for the selected mouse
single_mouse_data = capomulin_data[capomulin_data['Mouse ID'] == selected_mouse]

# Step 2: Generate a line plot of tumor volume versus time point for the selected mouse
plt.figure(figsize=(10, 6))
plt.plot(single_mouse_data['Timepoint'], single_mouse_data['Tumor Volume (mm3)'], marker='o', linestyle='-', color='blue')
plt.title(f'Tumor Volume vs. Time Point for Mouse {selected_mouse} (Capomulin)')
plt.xlabel('Time Point')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(True)
plt.show()

# Step 3: Generate a scatter plot of mouse weight versus average observed tumor volume for Capomulin regimen
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

plt.figure(figsize=(10, 6))
plt.scatter(mouse_weights, average_tumor_volume, color='green', edgecolors='black')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

```python
#6. Calculate Correlation and Regression

# Step 0: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Step 1: Extract relevant data for Capomulin regimen
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

# Step 2: Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(mouse_weights, average_tumor_volume)[0, 1]
print(f"Correlation Coefficient between Mouse Weight and Average Tumor Volume: {correlation_coefficient}")

# Step 3: Perform linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weights, average_tumor_volume)
regress_values = mouse_weights * slope + intercept
equation = f"y = {round(slope, 2)}x + {round(intercept, 2)}"

# Step 4: Plot the scatter plot and linear regression model
plt.figure(figsize=(10, 6))
plt.scatter(mouse_weights, average_tumor_volume, color='green', edgecolors='black')
plt.plot(mouse_weights, regress_values, "r-", label=equation)
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin) with Linear Regression')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.legend()
plt.grid(True)
plt.show()
```
